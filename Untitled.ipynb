```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("QVI_data.csv")
data.head()
```

```python
data.columns
```

```python
data["DATE"] = pd.to_datetime(data["DATE"])

data["MONTH_ID"] = (
    data["DATE"].dt.year * 100 +
    data["DATE"].dt.month
)

data[["DATE","MONTH_ID"]].head()
```

```python
monthly_metrics = data.groupby(
    ["STORE_NBR","MONTH_ID"]
).agg(
    TOT_SALES=("TOT_SALES","sum"),
    N_CUSTOMERS=("LYLTY_CARD_NBR","nunique"),
    N_TXNS=("TXN_ID","nunique")
).reset_index()

monthly_metrics.head()
```

```python
monthly_metrics[
    monthly_metrics["STORE_NBR"].isin([77,86,88])
].head(20)
```

```python
trial77 = monthly_metrics[
    monthly_metrics["STORE_NBR"] == 77
]

control233 = monthly_metrics[
    monthly_metrics["STORE_NBR"] == 233
]
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    trial77["MONTH_ID"],
    trial77["TOT_SALES"],
    marker="o",
    label="Trial Store 77"
)

plt.plot(
    control233["MONTH_ID"],
    control233["TOT_SALES"],
    marker="o",
    label="Control Store 233"
)

plt.title("Store 77 vs Control Store 233 - Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    trial77["MONTH_ID"],
    trial77["N_CUSTOMERS"],
    marker="o",
    label="Trial Store 77"
)

plt.plot(
    control233["MONTH_ID"],
    control233["N_CUSTOMERS"],
    marker="o",
    label="Control Store 233"
)

plt.title("Store 77 vs Control Store 233 - Customers")
plt.xlabel("Month")
plt.ylabel("Number of Customers")
plt.legend()
plt.grid(True)

plt.show()
```
