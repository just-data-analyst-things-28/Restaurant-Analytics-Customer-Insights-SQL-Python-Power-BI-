```python
import pandas as pd

zomato = pd.read_excel("/content/Copy of zomato_restaurants_in_India.xlsx")
swiggy = pd.read_excel("/content/Copy of swiggy.xlsx")
print(zomato.shape)
print(swiggy.shape)

print(zomato.columns.tolist())
print(swiggy.columns.tolist())
```

```python
zomato.info()
```

```python
print(zomato.columns.tolist())
```

```python
zomato.isnull().sum().sort_values(ascending=False)
```

```python
zomato["timings"] = zomato["timings"].fillna("Not Available")

zomato["cuisines"] = zomato["cuisines"].fillna("Unknown")
```

```python
zomato.isnull().sum().sum()
```

```python
# Drop columns with little analytical value

zomato = zomato.drop(
    columns=[
        'url',
        'address',
        'zipcode',
        'opentable_support'
    ],
    errors='ignore'
)

# Fill remaining missing values

zomato['cuisines'] = zomato['cuisines'].fillna('Unknown')

zomato['timings'] = zomato['timings'].fillna('Not Available')
```

```python
zomato.isnull().sum().sum()
```

```python
zomato["cost_category"] = pd.cut(
    zomato["average_cost_for_two"],
    bins=[0,500,1000,2000,100000],
    labels=["Budget","Mid Range","Premium","Luxury"]
)
```

```python
zomato["cost_category"].value_counts()
```

```python
zomato["popularity"] = pd.cut(
    zomato["votes"],
    bins=[-1,50,200,1000,1000000],
    labels=["Low","Medium","High","Very High"]
)
```

```python
zomato["popularity"].value_counts()
```

```python
zomato["aggregate_rating"].describe()
```

```python
swiggy.info()
```

```python
swiggy.isnull().sum()
```

```python
swiggy.columns = (
    swiggy.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
swiggy.columns
```

```python
swiggy["id"].duplicated().sum()
```

```python
swiggy.drop_duplicates(
    subset="id",
    inplace=True
)
```

```python
swiggy["price_category"] = pd.cut(
    swiggy["price"],
    bins=[0,200,400,700,10000],
    labels=[
        "Budget",
        "Mid Range",
        "Premium",
        "Luxury"
    ]
)
```

```python
swiggy["price_category"].value_counts()
```

```python
swiggy["delivery_category"] = pd.cut(
    swiggy["delivery_time"],
    bins=[0,30,45,60,500],
    labels=[
        "Fast",
        "Normal",
        "Slow",
        "Very Slow"
    ]
)
```

```python
swiggy["rating_category"] = pd.cut(
    swiggy["avg_ratings"],
    bins=[0,2,3,4,5],
    labels=[
        "Poor",
        "Average",
        "Good",
        "Excellent"
    ]
)
```

```python
zomato.to_csv(
    "zomato_clean.csv",
    index=False
)

swiggy.to_csv(
    "swiggy_clean.csv",
    index=False
)
```
